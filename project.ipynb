```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import the data and read the data
data= pd.read_csv("zomato.csv")
data.head()
```

```python
#drop unnecessary data
data=data.drop(['url','dish_liked','phone'],axis=1)
```

```python
data.info()
```

```python
data.head()
```

```python
data.duplicated().sum()
data.drop_duplicates(inplace=True)
```

```python
data.head(10)
```

```python
#calculate the NaN values from the dataset
data.isnull().sum()
```

```python
#Changing the Columns Names for easy way
data.columns
data = data.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type',
                                  'listed_in(city)':'city'})
```

```python
data.info()
```

```python
data.head(50)
```

```python
#trasforam the cost data for cost value becaused many time order cost in value 2 decimal value 
data['cost'] = data['cost'].astype(str)
data['cost'] = data['cost'].apply(lambda x: x.replace(',','.'))
data['cost'] = data['cost'].astype(float)
data.info()
```

```python
data.head(10)
```

```python
#Removing '/5' in rating for easy uderstanding Rates
data['rate'].unique()
data = data.loc[data.rate !='NEW']
data = data.loc[data.rate !='-'].reset_index(drop=True)
remove_slash = lambda x: x.replace('/5', '') if type(x) == np.str else x
data.rate = data.rate.apply(remove_slash).str.strip().astype('float')
data['rate'].head()
```

```python
data.head(10)
```

```python
# column names replace true and false for oder received and not and table booking and not booking
data.name = data.name.apply(lambda x:x.title())
data.online_order.replace(('Yes','No'),(True, False),inplace=True)
data.book_table.replace(('Yes','No'),(True, False),inplace=True)
```
